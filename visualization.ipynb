```
import folium
```

```
gcc_gdc["count"]=1
groupby=gcc_gdc.groupby(["Center Name_x"])[["count"]].sum().reset_index()
```

```
groupby.shape
```

```
merge=pd.merge(groupby,gcc_gdc[["Center Name_x","LONGITUDE","LATITUDE"]],how='left',on="Center Name_x")
```

```
merge_unique=merge.drop_duplicates(subset=['Center Name_x'], keep=False)
```

```
merge_unique.head()
```

```
merge_unique.shape
```

```
# New York latitude and longitude values
latitude = 40.730610
longitude = -73.935242
```

```
newyork_map = folium.Map(location=[latitude, longitude], zoom_start=12)

# display the map of New York
newyork_map
```

```
# instantiate a feature group for the incidents in the dataframe
incidents = folium.map.FeatureGroup()

# loop through violations and add each to the incidents feature group
for lat, lng, name,count in zip(merge_unique.LATITUDE, merge_unique.LONGITUDE,merge_unique["Center Name_x"],merge_unique["count"]):
    label = '{}, {}'.format(name,count)
    label = folium.Popup(label, parse_html=True)
    incidents.add_child(
        folium.Marker(
            [lat, lng],popup=label
        )
    )

# add incidents to map
newyork_map.add_child(incidents)
```
